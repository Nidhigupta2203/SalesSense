```python
import pandas as pd
import numpy as np

dates = pd.date_range(start="2022-01-01", end="2022-12-31")
sales = 120 + np.cumsum(np.random.normal(1.2, 3, len(dates)))

df = pd.DataFrame({
    "date": dates,
    "sales": sales.astype(int)
})

df.to_csv("../raw/sales.csv", index=False)
df.head()
```
